```python
import pandas as pd 
import seaborn as sns 
```

```python
tips=sns.load_dataset("tips")
```

```python
titanic=pd.read_csv("train.csv")
```

```python
flights=sns.load_dataset("flights")
```

```python
iris=sns.load_dataset("iris")
```

```python
# Scatterplot (Numerical - Numerical)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex",style="smoker",size="size")
```

```python
#bar plot (Numerical - Categorical)
```

```python
sns.barplot(x="Pclass",y="Fare",data=titanic,hue="Sex")
```

```python
# Box Plot (Numerical- Categorical)
```

```python
sns.boxplot(x="Sex",y="Age",data=titanic,hue="Survived")
```

```python
# distplot ( Numerical - Categorical)
sns.distplot(titanic[titanic["Survived"]==0]["Age"],hist=False)
sns.distplot(titanic[titanic["Survived"]==1]["Age"],hist=False)
```

```python
# Heat Map ( Categorical - Categorical )
```

```python
pt=pd.crosstab(titanic["Pclass"],titanic["Survived"])
#pivot table 
```

```python
sns.heatmap(pt)
```

```python
titanic.groupby("Pclass").mean()["Survived"]*100
```

```python
# Cluster Map 
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```

```python
# Pair Plot 
sns.pairplot(iris)
```

```python
# Line plot ( Numerical - Numerical )
sns.lineplot(x="month",y="Passengers",data=flights)
```
